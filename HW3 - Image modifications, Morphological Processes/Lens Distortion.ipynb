```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
import math as mt
```

```python
Row_Size = 712
Column_Size = 1072
Input_raw = open('classroom.raw','rb').read()
Input_raw = np.frombuffer(Input_raw,np.uint8)
Input_image = Input_raw[0:Row_Size*Column_Size]
Input_image = np.reshape(Input_image, (Row_Size,Column_Size))
plt.figure(num=1, figsize=(10, 10))
plt.matshow(Input_image,cmap='gray',fignum=1)
```

```python
Output_image = np.zeros( (Row_Size, Column_Size) ) 
fx = 600;
fy = 600;
k1 = -0.3536;
k2 = 0.1730;
k3 = 0;
xd = [];
yd = [];
uc = Row_Size/2;
vc = Column_Size/2;
for u in range (0,Row_Size):
    for v in range (0,Column_Size):
        x = ((u - uc)/fx);
        y = ((v - vc)/fy);
        r = (((x**2 + y**2)))**(1/2);
        xd = (x*( 1 + k1*(r**2) + k2*(r**4) + k3*(r**6)));
        yd =(y*( 1 + k1*(r**2) + k2*(r**4) + k3*(r**6)));
        convertedx = (int((xd*fx)+uc));
        convertedy = (int((yd*fy)+vc));
        Output_image[u][v] = Input_image[convertedx][convertedy];
    
```

```python
plt.figure(num=1, figsize=(10, 10))
plt.matshow(Output_image,cmap='gray',fignum=1)
```
